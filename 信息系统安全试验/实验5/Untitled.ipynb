```python
import random

def miller_rabin(n,k=5):
    if n<=1 or n==4:
        return False
    
    if n<=3:
        return True
    
    d=n-1
    while d%2==0:
        d//=2
    
    for _ in range(k):
        a=random.randint(2,n-2)
        x=pow(a,d,n)

        if x==1 or x==n-1:
            continue

        while d!=n-1:
            x=pow(x,2,n)
            d=d*2

            if x==n-1:
                break
        else:
            return False
    
    return True

def generate_large_prime(bits):
    while True:
        p=random.getrandbits(bits)
        if p%2==0:
            p=p+1

        p|=(1<<bits-1)|1

        if miller_rabin(p):
            return p

large_prime=generate_large_prime(2048)
print(large_prime)
```
