```python
import random
 
def is_prime(n):
    if n == 2:
        return True
    for i in range(2, int(n/2+1)):
        if n%i == 0:
            return False
    return True
 
#get prime p and q
def get_pq():
    p, q = 0, 0
    while True:
        p, q = random.randint(10, 100), random.randint(10, 100)
        if is_prime(int(p)) and is_prime(int(q)) and int(p)!=(q) :
            return int(p),int(q)
 
#get e and d
def get_ed(p,q):
    phn = (p-1)*(q-1)
    while True:
        e = random.randint(2, 50)
        if is_prime(e):
            for d in range(2, 1000):
                if e*d % phn == 1:
                    return e,d
 
p,q = get_pq()
n = p*q
e,d = get_ed(p,q)
print("public key: (n,e)=%d,%d" % (n,e))
print("private key: (n,d)=%d,%d" % (n,d))
m = random.randint(2, 100)
c = m**e%n
print("message m=%s" % m)
print("encrypt:\n c =%s" % c)
print("decrypt:\n m =%s" % (c**d%n))
```
